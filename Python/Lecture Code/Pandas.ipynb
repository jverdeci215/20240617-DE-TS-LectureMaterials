# Series
- 1D Array
- Indexed
- Basic Unit

```python
s = pd.Series([1,2,3,4,5])
s
```

```python
# We can utulize our own index
s = pd.Series([1,2,3], index=['a', 'b', 'c'])
s
```

```python
# Math operations on Series work based off the index
s1 = pd.Series([1,2,3], index=['a', 'b', 'c'])
s2 = pd.Series([4,5,6], index=['b', 'c', 'd'])

print(s1+s2)
```
