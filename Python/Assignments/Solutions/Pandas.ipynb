```python
path = 'Data/imdb_top_1000.csv'
```

```python
import pandas as pd

df = pd.read_csv(path)
```

```python
df.head()
```

```python
# Get rid of useless columns

df.drop(['Poster_Link', 'Certificate'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Oldest movie
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')
oldest = df.iloc[df['Released_Year'].idxmin()]
oldest
```

```python
# Newest Movie
newest = df.iloc[df['Released_Year'].idxmax()]
newest
```

```python
# Top 10 Movies
df.sort_values(['IMDB_Rating', 'Series_Title'], ascending=[False, True]).head(10)
```

```python
df['Genre'] = df['Genre'].str.split(', ')
df.head()
```

```python
df2 = df.explode('Genre')
df2.head()
```

```python
# Highest grossing movie for each genre
df['Genre'] = df['Genre'].str.split(', ')
df2 = df.explode('Genre')
df2.reset_index(drop=True, inplace=True)
df2.loc[df2.groupby(['Genre'])['IMDB_Rating'].idxmax()]
```

```python
# Star with the most movies
df3 = df[['Star1', 'Star2', 'Star3', 'Star4']]
df3.melt(var_name='columns', value_name='name').drop('columns', axis=1).value_counts()
```

```python
# Director with the most movies
df['Director'].value_counts()
```

```python
df['Gross'] = df['Gross'].str.replace(',', '', regex=False)
```

```python
df['Gross'] = df['Gross'].apply(pd.to_numeric, errors='coerce')
```

```python
df['Gross'] = df['Gross'].fillna(0)
```

```python
# Highest Grossing Movie by Director
df = df.dropna(subset=['Gross'])
df.loc[df.groupby("Director")['Gross'].idxmin()].sort_values('Gross', ascending=True).head(15)
```

```python
# Save the dataframe as parquet
df.to_parquet('Data/imdb.parquet')
```
